### Task 1 — Unit Converter

```python
def convert_length(value, direction):
    """
    Converts length between meters and feet.
    Parameters:
        value (float): The numeric value to convert.
        direction (str): 'mtf' for meters to feet, 'ftm' for feet to meters.
    Returns:
        float: The converted value.
    """
    if direction == 'mtf':
        return value * 3.28084
    elif direction == 'ftm':
        return value / 3.28084
    else:
        raise ValueError("Invalid direction. Use 'mtf' or 'ftm'.")


def convert_weight(value, direction):
    """
    Converts weight between kilograms and pounds.
    Parameters:
        value (float): The numeric value to convert.
        direction (str): 'ktl' for kg to lbs, 'ltk' for lbs to kg.
    Returns:
        float: The converted value.
    """
    if direction == 'ktl':
        return value * 2.20462
    elif direction == 'ltk':
        return value / 2.20462
    else:
        raise ValueError("Invalid direction. Use 'ktl' or 'ltk'.")


def convert_volume(value, direction):
    """
    Converts volume between liters and gallons.
    Parameters:
        value (float): The numeric value to convert.
        direction (str): 'ltg' for liters to gallons, 'gtl' for gallons to liters.
    Returns:
        float: The converted value.
    """
    if direction == 'ltg':
        return value * 0.264172
    elif direction == 'gtl':
        return value / 0.264172
    else:
        raise ValueError("Invalid direction. Use 'ltg' or 'gtl'.")


def unit_converter():
    """
    Prompts the user to select a conversion type, enter a value,
    and displays the converted result. Handles invalid input using try-except.
    """
    print("=" * 40)
    print("       UNIT CONVERTER")
    print("=" * 40)
    print("Choose conversion type:")
    print("1. Length  (meters <-> feet)")
    print("2. Weight  (kg <-> lbs)")
    print("3. Volume  (liters <-> gallons)")

    try:
        choice = input("Enter 1, 2, or 3: ").strip()

        if choice == '1':
            print("a. Meters to Feet\nb. Feet to Meters")
            sub = input("Enter a or b: ").strip().lower()
            value = float(input("Enter value: "))
            if sub == 'a':
                result = convert_length(value, 'mtf')
                print(f"\n{value} meters = {result:.4f} feet")
            elif sub == 'b':
                result = convert_length(value, 'ftm')
                print(f"\n{value} feet = {result:.4f} meters")
            else:
                print("Invalid sub-choice.")

        elif choice == '2':
            print("a. Kilograms to Pounds\nb. Pounds to Kilograms")
            sub = input("Enter a or b: ").strip().lower()
            value = float(input("Enter value: "))
            if sub == 'a':
                result = convert_weight(value, 'ktl')
                print(f"\n{value} kg = {result:.4f} lbs")
            elif sub == 'b':
                result = convert_weight(value, 'ltk')
                print(f"\n{value} lbs = {result:.4f} kg")
            else:
                print("Invalid sub-choice.")

        elif choice == '3':
            print("a. Liters to Gallons\nb. Gallons to Liters")
            sub = input("Enter a or b: ").strip().lower()
            value = float(input("Enter value: "))
            if sub == 'a':
                result = convert_volume(value, 'ltg')
                print(f"\n{value} liters = {result:.4f} gallons")
            elif sub == 'b':
                result = convert_volume(value, 'gtl')
                print(f"\n{value} gallons = {result:.4f} liters")
            else:
                print("Invalid sub-choice.")
        else:
            print("Unsupported conversion type. Please enter 1, 2, or 3.")

    except ValueError as e:
        print(f"Invalid input: {e}")


unit_converter()
```

### Task 2 — Mathematical Operations on a List

```python
def compute_sum(numbers):
    """
    Computes the sum of a list of numbers.
    Parameters:
        numbers (list): A list of numeric values.
    Returns:
        float: The sum of all elements.
    """
    total = 0
    for n in numbers:
        total += n
    return total


def compute_average(numbers):
    """
    Computes the average of a list of numbers.
    Parameters:
        numbers (list): A list of numeric values.
    Returns:
        float: The average (mean) of all elements.
    """
    if len(numbers) == 0:
        raise ValueError("Cannot compute average of an empty list.")
    return compute_sum(numbers) / len(numbers)


def compute_maximum(numbers):
    """
    Finds the maximum value in a list of numbers.
    Parameters:
        numbers (list): A list of numeric values.
    Returns:
        float: The maximum value.
    """
    if len(numbers) == 0:
        raise ValueError("Cannot find max of an empty list.")
    max_val = numbers[0]
    for n in numbers:
        if n > max_val:
            max_val = n
    return max_val


def compute_minimum(numbers):
    """
    Finds the minimum value in a list of numbers.
    Parameters:
        numbers (list): A list of numeric values.
    Returns:
        float: The minimum value.
    """
    if len(numbers) == 0:
        raise ValueError("Cannot find min of an empty list.")
    min_val = numbers[0]
    for n in numbers:
        if n < min_val:
            min_val = n
    return min_val


def math_operations():
    """
    Prompts user to choose a math operation and input a list of numbers.
    Performs the selected operation and displays the result.
    Handles errors using try-except.
    """
    print("=" * 40)
    print("   MATHEMATICAL LIST OPERATIONS")
    print("=" * 40)
    print("Choose an operation:")
    print("1. Sum\n2. Average\n3. Maximum\n4. Minimum")

    try:
        choice = input("Enter 1, 2, 3, or 4: ").strip()
        raw = input("Enter numbers separated by spaces: ").strip()

        if not raw:
            raise ValueError("Empty input. Please enter at least one number.")

        numbers = [float(x) for x in raw.split()]

        if choice == '1':
            print(f"\nSum = {compute_sum(numbers)}")
        elif choice == '2':
            print(f"\nAverage = {compute_average(numbers):.4f}")
        elif choice == '3':
            print(f"\nMaximum = {compute_maximum(numbers)}")
        elif choice == '4':
            print(f"\nMinimum = {compute_minimum(numbers)}")
        else:
            print("Invalid choice. Please enter 1, 2, 3, or 4.")

    except ValueError as e:
        print(f"Error: {e}")


math_operations()
```

## 4.2 Exercise on List Manipulation

```python
def extract_every_other(lst):
    """
    Returns every other element from lst starting from index 0.
    Parameters:
        lst (list): Input list.
    Returns:
        list: Every other element.
    """
    return lst[::2]

print("1. Extract Every Other Element:")
print(extract_every_other([1, 2, 3, 4, 5, 6]))
```

```python
def get_sublist(lst, start, end):
    """
    Returns a sublist from start to end index (inclusive).
    Parameters:
        lst (list): Input list.
        start (int): Starting index.
        end (int): Ending index (inclusive).
    Returns:
        list: Sublist from start to end.
    """
    return lst[start:end + 1]

print("2. Slice a Sublist:")
print(get_sublist([1, 2, 3, 4, 5, 6], 2, 4))
```

```python
def reverse_list(lst):
    """
    Returns a reversed version of lst using slicing.
    Parameters:
        lst (list): Input list.
    Returns:
        list: Reversed list.
    """
    return lst[::-1]

print("3. Reverse a List:")
print(reverse_list([1, 2, 3, 4, 5]))
```

```python
def remove_first_last(lst):
    """
    Returns lst without its first and last elements.
    Parameters:
        lst (list): Input list.
    Returns:
        list: List without first and last elements.
    """
    return lst[1:-1]

print("4. Remove First and Last:")
print(remove_first_last([1, 2, 3, 4, 5]))
```

```python
def get_first_n(lst, n):
    """
    Returns the first n elements of lst.
    Parameters:
        lst (list): Input list.
        n (int): Number of elements to extract.
    Returns:
        list: First n elements.
    """
    return lst[:n]

print("5. Get First n Elements:")
print(get_first_n([1, 2, 3, 4, 5], 3))
```

```python
def get_last_n(lst, n):
    """
    Returns the last n elements of lst.
    Parameters:
        lst (list): Input list.
        n (int): Number of elements to extract from the end.
    Returns:
        list: Last n elements.
    """
    return lst[-n:]

print("6. Get Last n Elements:")
print(get_last_n([1, 2, 3, 4, 5], 2))
```

```python
def reverse_skip(lst):
    """
    Returns every second element starting from second-to-last, moving backward.
    Parameters:
        lst (list): Input list.
    Returns:
        list: Elements in reverse order skipping one each time.
    """
    return lst[-2::-2]

print("7. Reverse Skip:")
print(reverse_skip([1, 2, 3, 4, 5, 6]))
```

## 4.3 Exercise on Nested Lists

```python
def flatten(lst):
    """
    Flattens a nested list into a single 1D list.
    Parameters:
        lst (list): A nested list.
    Returns:
        list: A flat list containing all elements.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result

print("1. Flatten:", flatten([[1, 2], [3, 4], [5]]))
```

```python
def access_nested_element(lst, indices):
    """
    Accesses an element in a nested list using a list of indices.
    Parameters:
        lst (list): A nested list.
        indices (list): A list of integer indices.
    Returns:
        The element at the specified position.
    """
    element = lst
    for i in indices:
        element = element[i]
    return element

lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print("2. Access Nested Element:", access_nested_element(lst, [1, 2]))
```

```python
def sum_nested(lst):
    """
    Recursively sums all elements in a nested list.
    Parameters:
        lst (list): A nested list of numbers.
    Returns:
        float: Sum of all elements.
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

print("3. Sum Nested:", sum_nested([[1, 2], [3, [4, 5]], 6]))
```

```python
def remove_element(lst, elem):
    """
    Removes all occurrences of elem from a nested list.
    Parameters:
        lst (list): A nested list.
        elem: The element to remove.
    Returns:
        list: Nested list with all occurrences of elem removed.
    """
    return [remove_element(item, elem) if isinstance(item, list)
            else item for item in lst if item != elem]

print("4. Remove Element:", remove_element([[1, 2], [3, 2], [4, 5]], 2))
```

```python
def find_max(lst):
    """
    Finds the maximum element in a nested list of any depth.
    Parameters:
        lst (list): A nested list of numbers.
    Returns:
        The maximum value found.
    """
    flat = flatten(lst)
    return max(flat)

print("5. Find Max:", find_max([[1, 2], [3, [4, 5]], 6]))
```

```python
def count_occurrences(lst, elem):
    """
    Counts how many times elem appears in a nested list.
    Parameters:
        lst (list): A nested list.
        elem: The element to count.
    Returns:
        int: Number of occurrences of elem.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count

print("6. Count Occurrences:", count_occurrences([[1, 2], [2, 3], [2, 4]], 2))
```

```python
def deep_flatten(lst):
    """
    Flattens a deeply nested list of any depth into a single list.
    Parameters:
        lst (list): A deeply nested list.
    Returns:
        list: A fully flattened single list.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

print("7. Deep Flatten:", deep_flatten([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]))
```

```python
def average_nested(lst):
    """
    Calculates the average of all elements in a nested list.
    Parameters:
        lst (list): A nested list of numbers.
    Returns:
        float: The average of all elements.
    """
    flat = flatten(lst)
    return sum(flat) / len(flat)

print("8. Average Nested:", average_nested([[1, 2], [3, 4], [5, 6]]))
```

### Problem 4 — Matrix Operations

```python
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

A_inv = np.linalg.inv(A)
identity_check = np.round(np.dot(A, A_inv))
print("1. A × A⁻¹ = I (rounded):\n", identity_check)

AB = np.dot(A, B)
BA = np.dot(B, A)
print("\n2. AB:\n", AB)
print("   BA:\n", BA)
print("   AB == BA?", np.array_equal(AB, BA))

AB_T = np.dot(A, B).T
BT_AT = np.dot(B.T, A.T)
print("\n3. (AB)^T:\n", AB_T)
print("   B^T × A^T:\n", BT_AT)
print("   Equal?", np.array_equal(AB_T, BT_AT))
```

```python
A_sys = np.array([[2, -3, 1],
                  [1, -1, 2],
                  [3,  1, -1]])

B_sys = np.array([-1, -3, 9])

A_inv_sys = np.linalg.inv(A_sys)
X_inverse = np.dot(A_inv_sys, B_sys)
print("Solution using Inverse Method:")
print(f"  x = {X_inverse[0]:.4f}, y = {X_inverse[1]:.4f}, z = {X_inverse[2]:.4f}")

X_solve = np.linalg.solve(A_sys, B_sys)
print("\nSolution using np.linalg.solve:")
print(f"  x = {X_solve[0]:.4f}, y = {X_solve[1]:.4f}, z = {X_solve[2]:.4f}")
```
